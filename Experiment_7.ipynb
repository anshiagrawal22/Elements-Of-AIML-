```python
import pandas as pd
df = pd.read_csv('/content/DATASET_COS.csv')
df.head()
```

```python
# Using StandardScaler for scaling numerical features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max scaling to numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = scaler.fit_transform(df[num_cols])

# Check the scaled data
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Apply standardization to numerical columns
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Applying one-hot encoding to categorical columns
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Use sparse_output instead of sparse
cat_cols = df.select_dtypes(include=['object']).columns  # Identify categorical columns
encoded_cols = pd.DataFrame(encoder.fit_transform(df[cat_cols]), columns=encoder.get_feature_names_out(cat_cols))

# Drop the original categorical columns and concatenate the encoded columns
df_encoded = df.drop(cat_cols, axis=1).reset_index(drop=True)
df_encoded = pd.concat([df_encoded, encoded_cols], axis=1)

# Display the updated dataframe
print(df_encoded.head())
```

```python
# Save the cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)
```
